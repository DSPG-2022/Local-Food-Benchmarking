```python
import requests
from bs4 import BeautifulSoup
```

```python
URL = "https://shop.iowafood.coop/product-search?IDSubcat=164"
page = requests.get(URL)

print(page.text)
```

```python
soup = BeautifulSoup(page.content, "lxml")
print(soup.prettify())
```

```python
item_name = soup.findAll('div',{'class':'row row-cols-1 row-cols-sm-2 row-cols-lg-3 no-gutters'})
print(item_name)
```

```python
#find elements by ID
product_name = soup.findAll('a',{'href':"/product/"})
print(product_name)

```
