```python
from bs4 import BeautifulSoup
import requests
import pandas as pd

html_text = requests.get("https://www.prudentproduce.net/shop/fruit").text
soup = BeautifulSoup(html_text, "lxml")
items = soup.find_all('div', class_ = 'span4 v2shopitem')

dfObj = pd.DataFrame(columns = ["Item Name", "Unit of Measurement", 'Price'])

for item in items:
  item_name = item.find('div', class_ = "smlholder").text.strip()
  item_unit = item.find("div", class_ = "pull-left shopunitsv2").text.strip()
  item_price = item.find("div", class_ = "pull-right shoppricesv3").text.replace(" ", "")
  item_price_list = item_price.split()
  item_price = item_price_list[1] + item_price_list[0]

  dfObj = dfObj.append({"Item Name": item_name, "Unit of Measurement": item_unit, "Price": item_price},
                       ignore_index = True)
```
